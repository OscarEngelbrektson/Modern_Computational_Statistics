```python
from scipy import stats

#Define f(x)
def f(x):
    return x**3 + 2*x +1

#Normal distribution with \[Mu] = 0 and \[Sigma] = 1
dist = stats.norm(0, 1)

samples = 100000
success_counter = 0

for i in range(samples):
    
    #Draw random sample x from normal distribution 
    x = dist.rvs()
    
    #If f(x) > 1, increment counter
    if f(x) > 1:
        success_counter += 1

print("Number of times f(x)>1 over 100 000 samples:", success_counter,
      "\n Fraction of samples x for which f(x) > 1:", success_counter/samples)
```
