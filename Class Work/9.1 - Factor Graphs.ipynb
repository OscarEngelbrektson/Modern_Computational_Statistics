```python
import scipy.stats as sts
import matplotlib.pyplot as plt
```

```python
#Fixed hyperparameters
alpha = 1
beta = 1

players = ["p1", "p2", "p3"]
player_scores = {}

#For each player
for i in range(3):
    #Player unique scoring probability
    p = sts.beta(alpha, beta).rvs()
    
    #Number of shots n in a game is  is drawn from N(10, 3)
    #Number of scores ~ Binomial(n, p)
    scores = [sts.binom(int(sts.norm(10, 3).rvs()), p).rvs() for _ in range(4)]
    
    #Record scoring
    player_scores[players[i]] = scores

for i in range(3):
    plt.hist(player_scores[players[i]], label=players[i], alpha=0.5)

plt.legend(loc=0)
plt.xlabel("Number of hoops")
plt.ylabel("Frequency (#games)")
plt.title("Number of hoops per game for 3 players over 4 games")
plt.show()
```
